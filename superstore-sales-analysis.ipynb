# **Data Description**
1. Row ID- Unique id of the row
1. Order ID- Customer Unique Id
1. Order Date- Date of order of product from the store
1. Ship Date- Date the product was shipped
1. Ship Mode- different shipping methods or delivery services available for sending a product to a customer,"Second Class" is one type of shipping mode.
1. Customer ID- Each customer unique id
1. Customer Name- Name of customer that order a product
1. Segment- Customer Segment
1. Country- Customer residence Country
1. City - Customer residence City
1. State- Customer residence State
1. Postal Code- Customer postal code
1. Region- Region of the Country where the Customer Reside
1. Product ID- Unique Product Identify
1. Category- Product Category in the store
1. Sub-Category- Product Subcategory in the Store
1. Product Name - Name of each product in the store
1. Sales- Sales of the product

```python
# cleaning 
df.isna().sum()
```

```python
df = df.dropna(subset=["Postal Code"])
df.isna().sum()
```

```python
df.duplicated()
```

```python
# rounding Postal Code column as it is not nessacary to include decimals
df["Postal Code"] = df["Postal Code"].round(0).astype("Int64")
df.head(10)
```

```python
# Reformat the 'Order Date' column
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=True, errors='coerce')
df['Ship Date'] = pd.to_datetime(df['Ship Date'], dayfirst=True, errors='coerce')

df.head(10)
```

```python
# Checking modified data types for some columns 
df.dtypes
```

# **Insights**
1. Which product category has the highest sales?
1. Which region generated the most sales?
1. What is the most preferred delivery method among customers?
1. Which cities rank in the top 10 by sales volume?
1. Which customer segment has the highest number of purchases?

```python
category_sales = df.groupby('Category')['Sales'].sum().sort_values()

# Create a bar plot
plt.figure(figsize=(8,6))
sns.barplot(x=category_sales.values, y=category_sales.index, palette="viridis")
plt.xlabel('Total Sales')
plt.ylabel('Category')
plt.title('Total Sales by Category')
plt.show()
```

```python
# Group by 'Region' and sum the 'Sales'
region_sales = df.groupby('Region')['Sales'].sum().sort_values()

# Create a bar plot
plt.figure(figsize=(8,6))
sns.barplot(x=region_sales.values, y=region_sales.index, palette="magma")
plt.xlabel('Total Sales')
plt.ylabel('Region')
plt.title('Total Sales by Region')
plt.show()
```

```python
ship_mode_counts = df['Ship Mode'].value_counts()

# Create a pie chart
plt.figure(figsize=(6,6))
plt.pie(ship_mode_counts, labels=ship_mode_counts.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title('Distribution of Ship Modes')
plt.show()
```

```python
top_cities = df.groupby('City')['Sales'].sum().sort_values(ascending=False).head(10)

# Create a bar plot
plt.figure(figsize=(10,6))
sns.barplot(x=top_cities.values, y=top_cities.index, palette="coolwarm")
plt.xlabel('Total Sales')
plt.ylabel('City')
plt.title('Top 10 Cities by Sales')
plt.show()
```

```python
segment_counts = df['Segment'].value_counts()

# Create a bar chart
plt.figure(figsize=(8, 5))
segment_counts.plot(kind='bar', color='lightcoral')

# Adding labels and title
plt.xlabel('Segment')
plt.ylabel('Number of Orders')
plt.title('Distribution of Customer Segments')

# Display the plot
plt.tight_layout()
plt.show()
```

# **Key Findinings:**
1. The Technology product category achieved the highest sales figures.
1. The West region emerged as the top performer in overall sales.
1. Standard Class delivery was the preferred choice among customers.
1. New York City ranked first in sales volume among the top 10 cities.
1. The Consumer customer segment recorded the highest number of purchases.
